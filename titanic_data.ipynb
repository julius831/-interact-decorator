```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv\n",)
```

```python
#distrubution of age
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['age'], kde=True)
plt.show()
```

```python
#survival by gender
sns.barplot(x='sex', y='survived', data=df)
plt.show()
```

```python
sns.countplot(x='sex', data=df)
plt.show()

```

```python
#females survived more
sns.barplot(x='sex', y='survived', data=df)
plt.show()
```

```python
#survival by age
sns.boxplot(x='survived', y='age', data=df)
plt.show()
```

```python
#survival embark town class
sns.catplot(x='embark_town', y='survived', hue='sex', col='class', kind='bar', data=df)
plt.show()
```
